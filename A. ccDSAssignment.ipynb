While the dependent variable must be binary, the independent variables can fall into various categories of data such as continuous data, discrete ordinal data, and discrete nominal data. So, if we are unsure whether to use logistic regression, we should ask ourselves if the dependent variable is dichotomous – fits into one of two set categories, and if the independent variables are either interval, ratio, or ordinal. </p>

#### Random Forest
 <p>The second machine learning algorithm that I will use is the random forest algorithm. As I have done for the logistic regression algorithm, I will do some research into the random forest algorithm to better understand it before I apply it to my current task. Random Forest is a powerful and versatile supervised machine learning algorithm that grows and combines multiple decisions trees to create a forest. It can be used for both classification and regression problems when coding with Python or R. Random forest tends to combine hundreds of decision trees, and then trains each decision tree on a different sample of observations. The final predictions of the random forest are made by averaging the predictions of each tree.</p>

<p>I have mentioned that it works with both classification and regression problems, and I am already familiar with classification which is when the dependent variable is categorical, but what is regression? I investigated it, and I found out that regression analysis is when the dependent variable is a real or continuous value such as salary, age, or weight. </p>

<p>To go into more detail with how the random forest algorithm works, it grows multiple decision trees, which are then merged for a more accurate prediction. The logic behind it is that multiple uncorrelated models, which are in this case the individual decision trees perform much better as a group than they do alone. When using the random forest model for classification, each tree gives a classification, and the forest chooses the classification with most of the votes. When using the random forest model for regression, the forest picks the average of the outputs of all trees. The key lies in the fact that there is a low correlation between the decision trees that make up the random forest model. While individual decision trees may produce errors, most of the group will be correct, moving the overall outcome in the right direction.</p>
<p>Random forests are used by data scientists in many industries such as banking, stock trading, medicine, and e-commence. It is used in banking to detect customers who are more likely to repay their debt on time, as well as to predict who will use the bank’s services frequently. It is used in stock trade to predict a stock’s future behaviour, and in healthcare to analyse a patient’s medical history to identify diseases. It is even used by scientists in China to study the spontaneous combustion patterns of coal to reduce the safety risks in coal mines. </p>

<p>What are the benefits of the random forest algorithm? It is versatile, able to be used for both regression and classification tasks. It is also considered an accurate and robust method because of the number of decision trees. One of the biggest problems in machine learning is overfitting, which refers to a model that models the training data too well, but with random forest, this usually will not happen if there are enough decision trees. </p>

<p>There are some drawbacks to the random forest algorithm, with the main one being that many trees can make the algorithm too slow and ineffective for real-time predictions. More trees give a more accurate prediction, but a slower model. Random forest is also a predictive modelling tool, not a descriptive one, meaning that other approaches would be better if you are looking for a description of the relationships in the data. </p>

<p>I think that the random forest algorithm will be suitable for me to use because my dependent variable is categorical, and the algorithm works with a categorical dependent variable, and I think that the accuracy that the algorithm provides because of its ability to use multiple decision trees will prove extremely beneficial to me while I am making my prediction. </p>

#### KNN
<p>The third and final machine learning algorithm that I will use is the KNN algorithm. As usual, I will do some research into this algorithm to ensure that I understand it before I start trying to apply the algorithm to my task. KNN is short for K-Nearest Neighbours and is a supervised learning algorithm that can be used for classification and regression problems but is most used for classification. It works on a principle assuming every data point falling in near to each other is falling in the same class, meaning that it classifies a new data point based on similarity.<p/>

<p>To make a prediction, the KNN algorithm will use the entire dataset. For an observation that is not part of the dataset and is the actual value we want to predict, the algorithm will look for K instances of the dataset closest to our observation. If KNN is used for a classification problem, the mode (most common value) of the dependent variables of the K closest observations will be used for predictions. If KNN is used for a regression problem, the mean (or median) of the y variables of the K closest observations will be used for predictions.<p/>

<p>What are the applications of the KNN algorithm? KNN is used for computer vision, finding, and predicting the most prominent image that resembles an input image. It is used in marketing by companies such as Amazon to recommend products to buy, and in content recommendation by companies such as Netflix to recommend movies to watch.<p/>

<p>What are the benefits of using the KNN model? Firstly, it is versatile, able to be used for classification, and regression, as well as search. It is simple and easy to implement. It is a lazy learning algorithm because it requires no training before making real-time predictions, making it faster than other algorithms that require training such as linear regression, and since the algorithm requires no training before making predictions, new data can be added seamlessly. There are only two parameters required to implement it, the value of K and the distance function.<p/>

<p>I will also need to consider the drawbacks of the KNN model. Accuracy depends heavily on the quality of the data. To use it effectively, we must find an optimal k value, and it is poor at classifying data points in a boundary where they can be classified one way or another. It also has a high prediction cost for large datasets because the cost of calculating the distance between the new point and each existing point becomes higher.<p/>

<p>I think that the KNN algorithm may be useful for me as it can be used for classification problems, but I am not sure if it will prove as useful as the previous two algorithms that I identified. Regardless, I would like to use it because I aim to use a variety of algorithms to see which one works best for me.<p/>

#### Explanation of the techniques that I used:
**Confusion Matrix:**<br /> 
A confusion matrix is a technique for summarizing the performance of a machine learning algorithm. Classification accuracy, which I describe below alone can be misleading if I have an unequal number of observations in each class, or if I have more than two classes in my dataset so it is best to also use a confusion matrix when evaluating my model. A confusion matrix will help me understand what my model is getting right, and what type of errors that it is making.

It uses a specific table layout with four different possible combinations of predicted and actual values. Please refer to the table below as I will use it to explain what each field in the table layout represents. True Positive represents how many times that I predicted positive (passed) and it was true. True Negative is how many times that I predicted negative (failed) and it was true. False Positive is how many times that I predicted positive (passed) and it was false. Finally, False Negative is how many times that I predicted negative (failed) and it was false. Knowing this information provides me with invaluable information about how well my model performed.

|                |                |
| -----------    | -----------    |
| True Positive  | False Positive |
| False Negative | True Negative  |

<br /> 

**Classification Report:**<br /> 
Classification reports are used to measure the quality of predictions from an algorithm. Some of its main features are described below.

**Precision:** Tells us what percentage of the predictions were correct. It describes the model's ability to avoid labelling an instance to be positive when it is negative.

The following formula is used to calculate precision:<br /> 
Precision = TP/(TP + FP)

**Recall:** Describes the model's ability to find positive instances.

The following formula is used to calculate recall:<br /> 
Recall = TP/(TP+FN)

**F1 score:** A-weighted harmonic mean of precision and recall. The weighted average of the F1 score is often used to compare models.

The following formula is used to calculate the F1 score:<br /> 
F1 Score = 2*(Recall * Precision) / (Recall + Precision)

**Support:** The number of actual occurrences of the class in the specified dataset. Imbalanced support may indicate structural weaknesses.

<br /> 

**Accuracy Score:**<br /> 
When we use the term accuracy, we are usually referring to classification accuracy. Classification accuracy is the ratio of the number of correct predictions to the total number of input samples.

Classification accuracy only works well if there are an equal number of samples belonging to each class. For example, if there were 98% samples of Class A and 2% samples of class B in a training set, the model can easily get 98% training accuracy by simply predicting every training sample belonging to class A. When the same model is tested on testing set with 60% samples of Class A, and 40% samples of class B, then the test accuracy would drop to 60%.

Classification accuracy is useful, but it can give us the false sense of achieving high accuracy. The real problem arises when the cost of misclassification of the minor class samples are very high. If we deal with a rare but fatal disease, the cost of failing to diagnose the disease of a sick person is much higher than the cost of sending a healthy person to more tests.
